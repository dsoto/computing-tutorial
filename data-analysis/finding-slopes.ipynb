# Finding slopes

Much of the data in energy comes from time series.  
We may want to know the slope between points to get the average power between to energy observations.


```python
import pandas as pd
import io

data_simple = '''date,energy (kWh)
2017-01-01 00:00,0.0
2017-01-01 01:00,1.5
2017-01-01 02:00,3.0
2017-01-01 03:00,4.5
2017-01-01 04:00,6.0
2017-01-01 05:00,7.5
2017-01-01 07:00,10.5
2017-01-01 08:00,12.0
'''

data = pd.read_csv(io.StringIO(data_simple), index_col=0, parse_dates=True)

data['time'] = data.index
```

```python
%matplotlib inline
data['energy (kWh)'].plot(marker='o', linestyle='')
```

We want to find the changes in energy and the changes in time.

$$ m = \frac{\Delta E}{\Delta t} $$

```python
import numpy as np
# note that the slope is for the hour or time period preceding the sample



delta_E = data['energy (kWh)'].diff(1)
delta_E.plot(marker='o', linestyle='')
```

```python
delta_T_hours = data['time'].diff(1) / np.timedelta64(1, 'h')
delta_T_hours.plot(marker='o', linestyle='')
```

```python
data['slopes'] = delta_E / delta_T_hours
data['slopes'].plot(marker='o')
# what are the units of this?
```
